```python
unique_values =  pd.unique(modelready['census_msa_cat'])
print(unique_values)
```

```python
m = RandomForestClassifier(n_jobs=-1)
m.fit(modelready.drop('h1n1_vaccine', axis=1), modelready.h1n1_vaccine)
m.score(modelready.drop('h1n1_vaccine', axis=1), modelready.h1n1_vaccine)
```

```python
# m.predict(testingmodelready)
results = pd.DataFrame(data=m.predict(testingmodelready))
print(results)
```
